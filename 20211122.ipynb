```python
import numpy as np
import IPython.display as ipd
```

```python
def phasor_sine(amp, sr, dur, freq, phase):
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta
```

```python
notes = [C, B, A, G, F, 0, G, A, C, B, A, G, F, E, 0, \
         A, G, F, E, D, 0, E, F, A, Ab, F, E, D, C4, 0, \
         Db, D, A, 0, A, 0, C, B, G, 0, \
         B3, C4, F, 0, F, 0, A, G, F, E, 0]

durs = [0.50, 0.50, 0.50, 0.25, 0.7, 0.25, 0.25, 0.45, 0.65, 0.45, 0.50, 0.25, 0.50, 1, 1.5, \
        0.40, 0.35, 0.50, 0.45, 0.75, 0.25, 0.35, 0.55, 0.30, 0.50, 0.45, 0.40, 0.50, 1, 1.5, \
        0.10, 0.40, 0.10, 0.05, 1.5, 1, 0.5, 0.5, 1.5, 2.5, \
        0.10, 0.40, 0.10, 0.05, 1.5, 1, 0.5, 0.5, 1.5, 2]
```

```python
s_all = np.empty(0)

for note, dur in zip(notes, durs):
  s, t, th= phasor_sine(amp, sr, dur, note, phase)
  s_all = np.concatenate((s_all, s))
```
